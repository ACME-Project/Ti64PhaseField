```python
G_Al_alpha = (Al_alpha_a[2] + Al_alpha_b[2]*T + Al_alpha_c[2]*T*np.log(T) + Al_alpha_d[2]*T*T) 
G_Al_beta = (Al_beta_a[2] + Al_beta_b[2]*T + Al_beta_c[2]*T*np.log(T) + Al_beta_d[2]*T*T) 

G_Ti_alpha = (Ti_alpha_a[1] + Ti_alpha_b[1]*T + Ti_alpha_c[1]*T*np.log(T) + Ti_alpha_d[1]*T*T) 
G_Ti_beta = (Ti_beta_a[0] + Ti_beta_b[0]*T + Ti_beta_c[0]*T*np.log(T) + Ti_beta_d[0]*T*T) 

G_V_alpha = (V_alpha_a[1] + V_alpha_b[1]*T + V_alpha_c[1]*T*np.log(T) + V_alpha_d[1]*T*T)  
G_V_beta = (V_beta_a[1] + V_beta_b[1]*T + V_beta_c[1]*T*np.log(T) + V_beta_d[1]*T*T) 

G_Alpha = ((cal*G_Al_alpha + cv*G_V_alpha + (1-cal-cv)*G_Ti_alpha +
             R*T*(cal*np.log(cal) + cv*np.log(cv) + (1-cal-cv)*np.log((1-cal-cv))) +
            cal*cv*L0_HCP_Al_V + cal*(1-cal-cv)*L0_HCP_Al_Ti + cv*(1-cal-cv)*L0_HCP_Ti_V))
 
G_Beta = (cal*G_Al_beta + cv*G_V_beta + (1-cal-cv)*G_Ti_beta +
             R*T*(cal*np.log(cal) + cv*np.log(cv) + (1-cal-cv)*np.log((1-cal-cv))) +
             cal*cv*L0_BCC_Al_V + cal*(1-cal-cv)*L0_BCC_Al_Ti + cv*(1-cal-cv)*L0_BCC_Ti_V +
             cal*cv*(1-cal-cv)*L0_BCC_Al_Ti_V)

G_Alpha = G_Alpha/(R*T)
G_Beta = G_Beta/(R*T)

print(G_Alpha, G_Beta)
```
